```python
#parts of speech tagging
import nltk
from nltk.corpus import wordnet
import string
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'): 
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
def clean_text(text):
    # lower text
    text = text.lower()
    sw_nltk = stopwords.words('english')
    sw_nltk=set(sw_nltk)
    set2={"don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"}
    stopword = set()
    stopword.update(sw_nltk - set2)
    words = [word for word in text.split() if word.lower() not in stopword]
    text = " ".join(words)
    # remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    #remove empty tokens
    text = [t for t in text if len(t) > 0]
    #pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in pos_tags]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.tokenize.treebank import TreebankWordDetokenizer
import networkx as nx
from nltk.corpus import stopwords

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')

def tokenize_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [token.lower() for token in tokens if token.isalpha() and token.lower() not in stop_words]
    
    return tokens

def build_graph(tokens, window_size=2):
    graph = nx.Graph()
    
    for i in range(len(tokens) - window_size + 1):
        for j in range(i + 1, i + window_size):
            if tokens[i] != tokens[j]:
                if graph.has_edge(tokens[i], tokens[j]):
                    graph[tokens[i]][tokens[j]]['weight'] += 1
                else:
                    graph.add_edge(tokens[i], tokens[j], weight=1)
    
    return graph

def textrank_keywords(text, num_keywords=5, window_size=2):
    tokens = tokenize_text(text)
    graph = build_graph(tokens, window_size)

    # Run PageRank algorithm
    scores = nx.pagerank(graph, weight='weight')

    # Get the top keywords
    keywords = sorted(scores, key=scores.get, reverse=True)[:num_keywords]

    return keywords

# Apply the textrank_keywords function to the 'review_clean' column
hotel['keywords'] = hotel['review_clean'].apply(textrank_keywords)

# Flatten the list of lists into a single list
flat_keywords = [keyword for sublist in hotel['keywords'] for keyword in sublist]

# Display the single list of keywords

print(flat_keywords)
```

```python
keyword_extracted = pd.DataFrame({'Keywords': flat_keywords})
```

```python
import pandas as pd

# Count the frequency of each keyword
keyword_counts = keyword_extracted['Keywords'].value_counts().reset_index()

# Rename the columns
keyword_counts.columns = ['Keywords', 'Frequency']

# Assign ranks based on frequency
keyword_counts['Rank'] = keyword_counts['Frequency'].rank(ascending=False)

# Sort the DataFrame by 'Rank' in ascending order
keyword_extracted_sorted = keyword_counts.sort_values(by='Rank', ascending=True)

# Select the top 10 keywords
top_10_keywords = keyword_extracted_sorted.head(10)

# Display the DataFrame with ranks and the top 10 keywords
print(keyword_extracted_sorted)
print("\nTop 10 Keywords with Highest Rank:")
print(top_10_keywords[['Keywords', 'Rank']])
```

```python
import pandas as pd
from textblob import TextBlob
 
def get_sentiment_score(text):
    analysis = TextBlob(str(text))
    return analysis.sentiment.polarity

# Apply the function to the 'review_clean' column and create a new column 'sentiment_score'
hotel['sentiment_score'] = hotel['review_clean'].apply(get_sentiment_score)

# Print the DataFrame with the new 'sentiment_score' column
print(hotel[['review_clean', 'sentiment_score','reviews.rating']])
```
