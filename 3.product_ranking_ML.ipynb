```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
  
hotel_reviews = data[data['name'] == 'Howard Johnson Inn - Newburgh']
 
reviews_room = hotel_reviews['room'].astype(str)
 
vectorizer_bow = CountVectorizer()
X_bow_room = vectorizer_bow.fit_transform(reviews_room)
 
tfidf_transformer = TfidfTransformer()
X_tfidf_room = tfidf_transformer.fit_transform(X_bow_room)
 
feature_names_tfidf = vectorizer_bow.get_feature_names_out()
 
X_tfidf_room_dense = X_tfidf_room.toarray()
 
df_tfidf_room = pd.DataFrame(X_tfidf_room_dense, columns=feature_names_tfidf)
 
sentiment_labels = hotel_reviews['sentiment_label']
 
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_room_dense, sentiment_labels, test_size=0.2, random_state=42)

nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
nb_predictions = nb_classifier.predict(X_test)
nb_confidences = nb_classifier.predict_proba(X_test)

lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)
lr_predictions = lr_classifier.predict(X_test)
lr_confidences = lr_classifier.predict_proba(X_test)

svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
svm_confidences = svm_classifier.decision_function(X_test)

print("Naive Bayes :",nb_confidences)
print("Linear Regression :",lr_confidences)
print("SVM :",svm_confidences)

print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_predictions))
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_predictions))
print("Support Vector Machine Accuracy:", accuracy_score(y_test, svm_predictions))
 
```
